# Feature Engineering

```python
df['Date_of_Journey'].str.split('/').str[0]
```

```python
df['Date'] = df['Date_of_Journey'].str.split('/').str[0]
df['Month'] =df['Date_of_Journey'].str.split('/').str[1]
df['Year'] = df['Date_of_Journey'].str.split('/').str[2]
```

```python
df.head()
```

```python
df['Date'] = df['Date'].astype(int)
df['Month'] = df['Month'].astype(int)
df['Year'] = df['Year'].astype(int)
```

```python
df.info()
```

```python
df.drop('Date_of_Journey',inplace=True,axis=1)
```

```python
df.head()
```

```python
df['Arrival_Time'] = df['Arrival_Time'].apply(lambda x:x.split(' ')[0])
```

```python
df.head()
```

```python
df['Arrival_hour'] = df['Arrival_Time'].apply(lambda x:x.split(':')[0])
df['Arrival_min'] = df['Arrival_Time'].apply(lambda x:x.split(':')[1])
```

```python
df['Arrival_hour'] = df['Arrival_hour'].astype(int)

df['Arrival_min'] = df['Arrival_min'].astype(int)
```

```python
df.drop('Arrival_Time',axis=1,inplace=True)
```

```python
df.head()
```

```python
df['Dep_hour'] = df['Dep_Time'].apply(lambda y : y.split(':')[0])
df['Dep_min'] = df['Dep_Time'].apply(lambda y : y.split(':')[1])
```

```python
df['Dep_hour'] = df['Dep_hour'].astype(int)

df['Dep_min'] = df['Dep_min'].astype(int)
```

```python
df.drop('Dep_Time',inplace=True,axis=1)
```

```python
df.head()
```

```python
df['Total_Stops'].unique()
```

```python
df['Total_Stops'] = df['Total_Stops'].replace({'non-stop': 0 ,'1 stop':1,'2 stops':2,'3 stops':3,'4 stops': 4, 'nan': 2})
```

```python
df.head()
```

```python
df.drop('Route',axis=1,inplace=True)
```

```python
df.head()
```

```python
df['Duration_hour'] = df['Duration'].apply(lambda y : y.split(' ')[0].split('h')[0])
```

```python
df.drop(6474,axis=0,inplace=True)
```

```python
df.drop(2660,axis=0,inplace=True)
```

```python
df['Duration_hour'] = df['Duration_hour'].astype(int)
```

```python
df.head()
```

```python
df['Duration_min'] = df['Duration'].apply(lambda y : y.split('m')[0].split('h')[1])
```

```python
df['Duration_min'] = df['Duration_min'].replace({'': 0})
```

```python
df['Duration_min']  = df['Duration_min'].astype(int)
```

```python
df.drop('Duration',axis=1,inplace=True)
```

```python
df.head()
```
